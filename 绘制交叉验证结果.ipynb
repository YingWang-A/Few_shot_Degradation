```python
%matplotlib inline
import math
import random
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import os
os.environ["KMP_DUPLICATE_LIB_OK"]  =  "TRUE"
```

```python
x = np.array(['0.8','0.9','1.0','1.1','1.2','1.3','1.4','1.5','1.6','1.7'])
y1 = np.array([0.0617,0.0609,0.0597,0.0603,0.0605,0.0608,0.0614,0.0619,0.0617,0.0624])
y2 = np.array([9.39,7.77,6.55,5.47,5.14,4.00,3.46,3.11,2.79,2.31])
```

```python
fig, ax1 = plt.subplots(figsize=(9,6))

color = 'tab:red'
ax1.set_xlabel(r"$\delta_1$", fontsize=20)
ax1.set_ylabel('Prediction error', color=color, fontsize=20)
ax1.plot(x, y1, color=color, linestyle='-', marker='o', markersize=10)
plt.ylim((0.0575, 0.0630))
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel('Deleted observation rate (%)', color=color, fontsize=20)
ax2.plot(x, y2, color=color, linestyle='--', marker='^', markersize=10)
plt.ylim((1.8, 10.00))
ax2.tick_params(axis='y', labelcolor=color)
plt.savefig('cross_validation_1.png', bbox_inches='tight', dpi=1200)
plt.show()
```
