```python
%matplotlib inline
import math
import random
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import os
os.environ["KMP_DUPLICATE_LIB_OK"]  =  "TRUE"
```

```python
x = np.array(['0.6','0.7','0.8','0.9','1.0','1.1','1.2','1.3','1.4'])
y1 = np.array([0.06382,0.06202,0.06147,0.06142,0.06048,0.06090,0.06120,0.06096,0.06131])
y2 = np.array([13.90,9.34,6.68,4.77,3.49,2.56,1.93,1.48,1.13])
```

```python
fig, ax1 = plt.subplots(figsize=(9,6))

color = 'tab:red'
ax1.set_xlabel(r"$\delta_1$", fontsize=20)
ax1.set_ylabel('Prediction error', color=color, fontsize=20)
ax1.plot(x, y1, color=color, linestyle='-', marker='o', markersize=10)
plt.ylim((0.0580, 0.0660))
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel('Deleted observation rate (%)', color=color, fontsize=20)
ax2.plot(x, y2, color=color, linestyle='--', marker='^', markersize=10)
plt.ylim((-1.0, 15.0))
ax2.tick_params(axis='y', labelcolor=color)
plt.savefig('cross_validation_1.png', bbox_inches='tight', dpi=1200)
plt.show()
```
