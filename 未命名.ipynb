```python
import csv
```

```python
dictLabels = {}
csvf = 'train.csv'
with open(csvf) as csvfile:
    csvreader = csv.reader(csvfile, delimiter=',')
    next(csvreader, None)
    for i, row in enumerate(csvreader):
        filename = row[0]
        label = row[1]
        if label in dictLabels.keys():
            dictLabels[label].append(filename)
        else:
            dictLabels[label] = [filename]
```
