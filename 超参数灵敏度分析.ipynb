```python
fig, [ax1, ax2, ax3, ax4] = plt.subplots(1,4,figsize=(20,4.4),layout="constrained")

#plt.subplot(141)
color = 'tab:red'
ax1.plot(x1,y1, color=color, marker='o', markersize=15)
ax1.set_xlabel('Number of hidden neurons', fontsize=20)
ax1.set_ylabel('RMSE', fontsize=20)

#plt.subplot(142)
ax2.plot(x2,y2, color=color, marker='o', markersize=15)
ax2.set_xlabel(r"$\lambda$", fontsize=20)
ax2.set_ylabel('RMSE', fontsize=20)

#plt.subplot(143)
color = 'tab:red'
ax3.set_xlabel(r"$M$", fontsize=20)
ax3.set_ylabel('RMSE', color=color, fontsize=20)
ax3.plot(x3, y3, color=color, linestyle='-', marker='o', markersize=15)
ax3.set_ylim((15.03, 15.38))
# ax3.tick_params(axis='y', labelcolor=color)

ax32 = ax3.twinx()
color = 'tab:blue'
ax32.set_ylabel('Inference time/s', color=color, fontsize=20)
ax32.plot(x3, z3, color=color, linestyle='--', marker='^', markersize=15)
ax32.set_ylim((-0.08, 1.98))
ax32.tick_params(axis='y', labelcolor=color)

# plt.subplot(144)
ax4.plot(x4,y4)

color = 'tab:red'
ax4.set_xlabel(r"$N$", fontsize=20)
ax4.set_ylabel('RMSE', color=color, fontsize=20)
ax4.plot(x4, y4, color=color, linestyle='-', marker='o', markersize=15)
ax4.set_ylim((15.085, 15.18))

ax42 = ax4.twinx()
color = 'tab:blue'
ax42.set_ylabel('Training time/s', color=color, fontsize=20)
ax42.plot(x4, z4, color=color, linestyle='--', marker='^', markersize=15)
ax42.set_ylim((0, 0.9))
ax42.tick_params(axis='y', labelcolor=color)

# plt.subplots_adjust(wspace = 0.5, hspace = 0.35)
plt.savefig('Hyperparameter sensitive.png', bbox_inches='tight', dpi=1200)
plt.show()
```
