```python
import math
import torch
from torch import nn
from torch.autograd import Variable
from torch.distributions import Normal
from torch.utils.data import DataLoader
from torch.nn.utils.rnn import pad_sequence
from torch.utils.data import Dataset
import torch.nn.functional as F
import torch.optim as optim
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from copy import deepcopy
%matplotlib inline
import os
os.environ["KMP_DUPLICATE_LIB_OK"]  =  "TRUE"
```

```python
#  读入传感器数据
df_train = pd.read_csv('train_data.csv')
num_units = 100

targets = []

for i in range(num_units):
    sensor_data = torch.tensor(df_train[df_train['Unit'] == i+1].iloc[:,2:].values, dtype = torch.float)  
    targets.append(sensor_data)

# 记录传感器数据的真实长度
lengths = []
for i in range(num_units):
    lengths.append(targets[i].shape[0])
    
# 对不等长的传感器数据进行填充
pad_targets = pad_sequence(targets, batch_first=True, padding_value=0.0)
print(pad_targets.shape)
```

```python
class Unit(Dataset):
    def __init__(self, batchsz, data, lengths):
        self.batchsz = batchsz
        self.data = data
        self.lengths = lengths
        
        self.num_units = len(data)
        self.creat_batch(self.batchsz)
    
    def creat_batch(self, batchsz):
        self.support_x_batch = []
        self.support_len_batch = []
        
        self.query_x_batch = []
        self.query_len_batch = []
        
        for b in range(batchsz):   # 采样batchsz个task
            selected_unit = np.random.choice(self.num_units, 1, False)[0]
            
            valid_len = self.lengths[selected_unit]
            
            query_x = self.data[selected_unit]
            support_x = self.data[selected_unit]
            query_len = valid_len            
            
            a = 500
            while a > valid_len:
                a = int(np.random.uniform(31,303))
            support_len = a
            
            self.support_x_batch.append(support_x)
            self.query_x_batch.append(query_x)
            
            self.support_len_batch.append(support_len)
            self.query_len_batch.append(query_len)
            
    def __getitem__(self, index):
        return self.support_x_batch[index], self.query_x_batch[index], self.support_len_batch[index], self.query_len_batch[index]
    
    def __len__(self):
        return self.batchsz
```

```python
batchsz = 100

train_data = pad_targets[:20]
train_lens = lengths[:20]

test_data = pad_targets[20:40]
test_lens = lengths[20:40]

train_unit = Unit(batchsz, train_data, train_lens)
test_unit = Unit(batchsz, test_data, test_lens)
```

```python
num_task = 10
train_loader = DataLoader(train_unit, num_task, shuffle=True)
test_loader = DataLoader(test_unit, 1, shuffle=True)
```

```python
for batch_idx, batch_units in enumerate(train_loader):
    support_x_batch, query_x_batch, support_len_batch, query_len_batch = batch_units
    print(support_x_batch.shape, query_x_batch.shape)
```

```python
class LSTM():
    def __init__(self, num_hiddens, num_outputs):
        self.Psi = torch.load('Psi.params')
        self.gamma = Variable(nn.Parameter(torch.randn(num_hiddens, num_outputs))) * 0.01
        self.gamma.requires_grad_(True)
        
    def __call__(self, gamma=None):
        if gamma is None:
            gamma = self.gamma
        return self.Psi @ gamma
```

```python
class Meta(nn.Module):
    def __init__(self, update_lr, meta_lr, num_task, update_step, update_step_test):
        super(Meta, self).__init__()
        
        self.update_lr = update_lr
        self.meta_lr = meta_lr
        self.num_task = num_task
        self.update_step = update_step
        self.update_test_step = update_step_test
        
        self.net = LSTM(3,6)
        self.meta_optim = optim.Adam([self.net.gamma], lr=self.meta_lr)
        
        self.mse_loss = nn.MSELoss()
        
    def forward(self, x_spt, x_qry, len_spt, len_qry):
        num_task = x_spt.size(0)
        
        losses_q = [0 for _ in range(self.update_step + 1)]
        
        for i in range(num_task):
            pred_y = self.net(self.net.gamma)[:len_spt[i]]
            true_y = x_spt[i][:len_spt[i]]
            
            loss = self.mse_loss(pred_y, true_y)
            grad = torch.autograd.grad(loss, self.net.gamma)
            fast_weights = list(map(lambda p: p[1] - self.update_lr * p[0], zip(grad, self.net.gamma)))[0]
            
            # this is the loss before first update
            with torch.no_grad():
                pred_y = self.net(self.net.gamma)[:len_qry[i]]
                true_y = x_qry[i][:len_qry[i]]
                loss_q = self.mse_loss(pred_y, true_y)
                losses_q[0] += loss_q
             
            # this is the loss after the first update
            with torch.no_grad():
                pred_y = self.net(fast_weights)[:len_qry[i]]
                true_y = x_qry[i][:len_qry[i]]
                loss_q = self.mse_loss(pred_y, true_y)
                losses_q[1] += loss_q
                
            for k in range(1, self.update_step):
                pred_y = self.net(fast_weights)[:len_spt[i]]
                true_y = x_spt[i][:len_spt[i]]
                loss = self.mse_loss(pred_y, true_y)
                grad = torch.autograd.grad(loss, fast_weights)
                fast_weights = list(map(lambda p: p[1] - self.update_lr * p[0], zip(grad, self.net.gamma)))[0]
                
                pred_y = self.net(fast_weights)[:len_qry[i]]
                true_y = x_qry[i][:len_qry[i]]
                loss_q = self.mse_loss(pred_y, true_y)
                losses_q[k+1] += loss_q
                
        loss_q = losses_q[-1] / num_task
          
        self.meta_optim.zero_grad()
        loss_q.backward()
        self.meta_optim.step() 
        
        return loss_q.item()
        
    def finetunning(self, x_spt, x_qry, len_spt, len_qry, test_epochs=None):
        if test_epochs is None:
            test_epochs = self.update_test_step
            
        querysz = x_qry.size(0)
        
        losses = [0 for _ in range(test_epochs + 1)]
        
        net = deepcopy(self.net)
        
        pred_y = net()[:len_spt]
        true_y = x_spt[:len_spt]
        loss = self.mse_loss(pred_y, true_y)
        grad = torch.autograd.grad(loss, net.gamma)
        fast_weights = list(map(lambda p: p[1] - self.update_lr * p[0], zip(grad, self.net.gamma)))[0]
        
        with torch.no_grad():
            pred_y = net()[:len_qry]
            true_y = x_qry[:len_qry]
            loss_q = self.mse_loss(pred_y, true_y) 
            losses[0] = losses[0] + loss_q
            
        with torch.no_grad():
            pred_y = net(fast_weights)[:len_qry]
            true_y = x_qry[:len_qry]
            loss_q = self.mse_loss(pred_y, true_y) 
            losses[1] = losses[1] + loss_q
        
        for k in range(1, test_epochs):
            pred_y = net(fast_weights)[:len_spt]
            true_y = x_spt[:len_spt]
            loss = self.mse_loss(pred_y, true_y)
            grad = torch.autograd.grad(loss, fast_weights)
            fast_weights = list(map(lambda p: p[1] - self.update_lr * p[0], zip(grad, fast_weights)))[0]
            
            with torch.no_grad():
                pred_y = net(fast_weights)[:len_qry]
                true_y = x_qry[:len_qry]
                loss_q = self.mse_loss(pred_y, true_y) 
                losses[k+1] = losses[k+1] + loss_q
        
        del net
        
        return fast_weights, losses      
```

```python
update_lr, meta_lr = 0.1, 0.001
update_step, update_step_test = 50, 100
maml = Meta(update_lr, meta_lr, num_task, update_step, update_step_test)
```

```python
torch.save(maml.net, 'MAML.pth')
```

```python
num_epochs = 200
for epoch in range(num_epochs):
    # fetch meta_batchsz num of episode each time
    db = DataLoader(train_unit, num_task, shuffle=True)
    print('epoch:{}/{}'.format(epoch+1, num_epochs))

    for step, (x_spt, x_qry, len_spt, len_qry) in enumerate(db):

        train_loss = maml(x_spt, x_qry, len_spt, len_qry)
        if step % 3 == 0:
            print(f'batch {step},loss{train_loss:f}')

        if (step+1) % 10 == 0:  # evaluation
            db_test = DataLoader(test_unit, 1, shuffle=True)
            test_all_loss = []

            for x_spt, x_qry, len_spt, len_qry in db_test:

                _, test_unit_loss = maml.finetunning(x_spt[0], x_qry[0], len_spt, len_qry)
                test_all_loss.append(test_unit_loss)

            # [b, update_step+1]
            test_loss = np.array(test_all_loss).mean(axis=0).astype(np.float16)
            print('Test loss:', test_loss[-5:])
```
